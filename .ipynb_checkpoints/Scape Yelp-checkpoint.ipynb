```python
import csv
import requests
from bs4 import BeautifulSoup
import os.path
import pandas as pd
from random import uniform
from time import sleep

def get_soup(url):
    try:
        sleep(uniform(.5, 10))
        response = requests.get(url, timeout=10)
        soup = BeautifulSoup(response.content, 'html.parser')
    except:
        soup = get_soup(url)
        
    return soup
```
